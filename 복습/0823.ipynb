```python
# 감소하는 수
from itertools import combinations

n = int(input())
nums = list()

for i in range(1, 11):
    for comb in combinations(range(0, 10), i):
        comb = list(comb)
        comb.sort(reverse=True)
        nums.append(int("".join(map(str, comb))))

nums.sort()

try:
    print(nums[n])
except:
    print(-1)
```

```python
# A -> B
a, b = map(int, input().split())

count = 1

while b != a:
    count += 1
    temp = b

    if b % 10 == 1:
        b //= 10
    elif b % 2 == 0:
        b //= 2
    
    if temp == b:
        print(-1)
        break
else:
    print(count)
```

```python
# isharp

string = input()

# ';' 제거 
string = string.rstrip(';')

string_list = []
# 공백 기준 배열에 저장
string_list += string.split(' ')
# basic type -> int& 저장 
basic_type = string_list[0]
# int& 배열에서 제거
del string_list[0]
# ,와 " " 제거
for string in string_list:
    string = string.replace(',', '').replace(' ', '')
    # basic type 출력
    print(basic_type, end="")
    # 반대로 출력
    for i in range(len(string) - 1, 0, -1):
        if not string[i].isalpha():
            if string[i] == ']':
                print("[", end="") 
            elif string[i] == '[':
                print("]",end="")
            else:
                print(string[i], end="")

    print(' ', end='')

    for i in range(len(string)):
        if string[i].isalpha():
            print(string[i], end='')   
    print(';')
        
# 각 배경 돌면서 -> print()
# [이면 print("]")
# ]이면 print("[")
# 마지막 ";"
# 띄어쓰기
```

```python
# 소수 찾기
n = int(input())
numbers = list(map(int, input().split()))

result = 0
for num in numbers:
    cnt = 0
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                cnt += 1
        if cnt == 0:
            result += 1
print(result)    
```

```python
# 소수
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

m = int(input())
n = int(input())

result = []
for i in range(m, n + 1):
    if is_prime(i):
        result.append(i)

if result:
    print(sum(result))
    print(min(result))
else:
    print(-1)
```

```python
# DFS와 BFS
from collections import deque

def dfs(graph, visited, start):
    visited[start] = True
    print(start, end=" ")

    for k in graph[start]:
        if not visited[k]:
            dfs(graph, visited, k)

def bfs(graph, visited, start):
    queue = deque([start])
    visited[start] = True

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for k in graph[node]:
            if not visited[k]:
                visited[k] = True
                queue.append(k)

n, m, v = map(int, input().split())

graph = [[] for _ in range(n+1)]

for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

for k in graph:
    k.sort()

visited_dfs = [False] * (n + 1)
visited_bfs = [False] * (n + 1)

dfs(graph, visited_dfs, v)
print()
bfs(graph, visited_bfs, v)
```
