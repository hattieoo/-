````python
# ÎÇ¥ÏΩîÎìú
from collections import deque
import math

n, m = map(int, input().split())

graph = [list(input()) for _ in range(m)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

visited_W = [[False] * n for _ in range(m)]
visited_B = [[False] * n for _ in range(m)]

def bfs(graph, visited, start, v):
    a, b = start[0], start[1]
    queue = deque([(a, b)])
    visited[a][b] = True

    if graph[a][b] == v:
        count = 1
    else:
        count = 0

    while queue:
        x, y = queue.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0 <= nx < n and 0 <= ny < m and not visited[nx][ny]:
                if graph[nx][ny] == v:    
                    queue.append((nx, ny))
                    visited[nx][ny] = True
                    count += 1                

    return count

result_W = []
result_B = []

for i in range(n):
    for j in range(m):
        if not visited_W[i][j]:
            result_W.append(bfs(graph, visited_W, (i, j), 'W'))
        if not visited_B[i][j]:
            result_B.append(bfs(graph, visited_B, (i, j), 'B')) 
sum_W = 0
sum_B = 0

for w, b in zip (result_W, result_B):
    if w != 0:
        sum_W += math.pow(w, 2)
    if b != 0:
        sum_B += math.pow(b, 2)

print(round(sum_W), round(sum_B))
````

````python
# Îã§Î•∏ ÏΩîÎìú
from collections import deque

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(graph, x, y, visited, N, M):
    queue = deque([(x, y)])
    visited[x][y] = True
    cnt = 1
    
    while queue:
        cur_x, cur_y = queue.popleft()
        
        for i in range(4):
            nx = cur_x + dx[i]
            ny = cur_y + dy[i]
            
            if 0 <= nx < N and 0 <= ny < M and not visited[nx][ny] and graph[nx][ny] == graph[x][y]:
                queue.append((nx, ny))
                visited[nx][ny] = True
                cnt += 1
                
    return cnt

N, M = map(int, input().split())
graph = [list(input()) for _ in range(M)]

visited = [[False] * N for _ in range(M)]
w_score = 0
b_score = 0

for i in range(M):
    for j in range(N):
        if not visited[i][j]:
            cnt = bfs(graph, i, j, visited, M, N)
            if graph[i][j] == 'W':
                w_score += cnt ** 2
            else:
                b_score += cnt ** 2

print(w_score, b_score)
````
